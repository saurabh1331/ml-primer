## What is Machine Learning 

You can think of machine learning as a pettern recognition. 

Essentially, it is a method of teaching computers to make and improve predictions or behaviors based on some data. What is this "data"? Well, that depends entirely on the problem. It could be readings from a robot's sensors as it learns to walk, or the correct output of a program for certain input.

Another way to think about machine learning is that it is "pattern recognition" - the act of teaching a program to react to or recognize patterns.

**Data Format**

Most of the data in machine learning can be represented in tables. Rows in the table correspond to data point. Columns in the tables corresponds to feature. Total number of feature is called Dimensionality of the data (D). So, every data point would have D number of data point. As a convention, Input is called X and output is called Y (labels). 

** Types of Machine learning **

There is mainly two types of machine learning.
- *Supervised Learning* : You have some pre knowledge of the data i.e. For some given x, y is known. Based on these data, algorithm learns the pattern and predict y for any newcoming x.

- *Unsupervised Learning* : You do not have pre labels to the data. We want to find out hidden structure in the data. 





**Example :**
- predicting the species of iris given a set of measurements of its flower.
- given a multicolor image of an object through a telescope, determine whether that object is a star, a quasar, or a galaxy.
- Given a photo, identify the person in the photo.

- Given the previous years sales of a store, predict the sales of next year. 
- Given the previous years weather data, predict how much would it rain this year?



So, supervised learning can be broken down into two categories.

 - Classification : Labels are discrete.

 - Regression: Labels are continuous.

```python
import pandas as pd
d = {'10th Math' : pd.Series([79,80,83,85], index=['John', 'mary', 'David', 'Lisa']),
     '10th History' : pd.Series([90,81,83,82], index=['John', 'mary', 'David', 'Lisa']),
     '12th Math' : pd.Series([72,80,71,89], index=['John', 'mary', 'David', 'Lisa']),
     '12th Physics' : pd.Series([81,83,85,87], index=['John', 'mary', 'David', 'Lisa']),
     'College Computers' : pd.Series([89,72,78,75], index=['John', 'mary', 'David', 'Lisa'])}
df_input = pd.DataFrame(d)
print df_input
```

```python
d = {'Success' : pd.Series([1,0,0,1], index=['a', 'b', 'c','d'])}
df_label = pd.DataFrame(d)
print df_label
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(C=1)
logreg.fit(df_input,df_label)
```

```python
print logreg.coef_
print df_input.columns
```
